```python
import matplotlib.pyplot as plt
# line 1 points
x1 = [16,32,64,100,144]
y1 = [9312,36960,147552,295000,589920]
# plotting the line 1 points
plt.plot(x1, y1, label = "Dictationary")

# line 2 points
x2 = [16,32,64,100,144]
y2 = [2200,6136,33048,85176,173016]
# plotting the line 2 points
plt.plot(x2, y2, label = "List")

# line 3 points
x3 = [16,32,64,100,144]
y3 = [2088,6040,32808,80040,165928]
# plotting the line 2 points
plt.plot(x3, y3, label = "Tuple")

# naming the x axis
plt.xlabel('No of Collection m')
# naming the y axis
plt.ylabel('Memory usage in Bytes')
# giving a title to my graph
plt.title('Memory usage of Collection versus No. of Collection m !')

# show a legend on the plot
plt.legend()

# function to show the plot
plt.show()
```
